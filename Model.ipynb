```python
df["t-4dev"] = df.groupby("DMA").dev.shift(-4)
df["t-8dev"] = df.groupby("DMA").dev.shift(-8)
df["t-12dev"] = df.groupby("DMA").dev.shift(-12)
df["t-16dev"] = df.groupby("DMA").dev.shift(-16)
df["t-20dev"] = df.groupby("DMA").dev.shift(-20)
df["t-24dev"] = df.groupby("DMA").dev.shift(-24)
df["t-28dev"] = df.groupby("DMA").dev.shift(-28)
df["t-32dev"] = df.groupby("DMA").dev.shift(-32)
df["t-36dev"] = df.groupby("DMA").dev.shift(-36)
df["t-40dev"] = df.groupby("DMA").dev.shift(-40)

df["t4Leak"] = df.groupby("DMA").exist.shift(4)
df["t8Leak"] = df.groupby("DMA").exist.shift(8)
df["t12Leak"] = df.groupby("DMA").exist.shift(12)
df["t16Leak"] = df.groupby("DMA").exist.shift(16)
df["t20Leak"] = df.groupby("DMA").exist.shift(20)
df["t24Leak"] = df.groupby("DMA").exist.shift(24)
df["t28Leak"] = df.groupby("DMA").exist.shift(28)
df["t32Leak"] = df.groupby("DMA").exist.shift(32)
df["t36Leak"] = df.groupby("DMA").exist.shift(36)
df["t40Leak"] = df.groupby("DMA").exist.shift(40)
```

```python
df.head()
```

```python
data = df[["DMA","Flow", "dev","t-4dev","t-8dev","t-12dev","t-16dev","t-20dev","t-24dev","t-28dev","t-32dev","t-36dev","t-40dev","t4Leak","t8Leak","t12Leak","t16Leak","t20Leak","t24Leak","t28Leak","t32Leak","t36Leak","t40Leak","exist"]]
data.head(100)
```

```python
data.isnull().sum()
```

```python
data.shape[0]
```

```python
data = data.dropna(subset=["t-4dev","t-8dev","t-12dev","t-16dev","t-20dev","t-24dev","t-28dev","t-32dev","t-36dev","t-40dev"], how='any') 
data["t4Leak"].fillna(False, inplace=True)
data["t8Leak"].fillna(False, inplace=True)
data["t12Leak"].fillna(False, inplace=True)
data["t16Leak"].fillna(False, inplace=True)
data["t20Leak"].fillna(False, inplace=True)
data["t24Leak"].fillna(False, inplace=True)
data["t28Leak"].fillna(False, inplace=True)
data["t32Leak"].fillna(False, inplace=True)
data["t36Leak"].fillna(False, inplace=True)
data["t40Leak"].fillna(False, inplace=True)
data.isnull().sum()
```

```python
data.shape[0]
```

```python
data["LeakFixIn4Days"] = data[["exist", "t4Leak","t8Leak","t12Leak","t16Leak","t20Leak","t24Leak","t28Leak","t32Leak","t36Leak","t40Leak"]].any(1)
```
