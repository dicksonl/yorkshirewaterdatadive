```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_pickle("./northnf")
df['dev'] = df.Flow - df.intervalMean  
```

```python
df["t-4dev"] = df.groupby("DMA").dev.shift(-4)
df["t-8dev"] = df.groupby("DMA").dev.shift(-8)
df["t-12dev"] = df.groupby("DMA").dev.shift(-12)
df["t-16dev"] = df.groupby("DMA").dev.shift(-16)

df["t+4Leak"] = df.groupby("DMA").exist.shift(4)
df["t+8Leak"] = df.groupby("DMA").exist.shift(8)
df["t+12Leak"] = df.groupby("DMA").exist.shift(12)
df["t+16Leak"] = df.groupby("DMA").exist.shift(16)
```

```python
df.head()
```

```python
data = df[["Flow", "dev","t-4dev","t-8dev","t-12dev","t-16dev","t+4Leak","t+8Leak","t+12Leak","t+16Leak","exist"]]
data.head(100)
```

```python
data.isnull().sum()
```

```python
data.shape[0]
```

```python
data = data.dropna(subset=['t-4dev', 't-8dev', 't-12dev', 't-16dev'], how='any') 
data["t+4Leak"].fillna(False, inplace=True)
data["t+8Leak"].fillna(False, inplace=True)
data["t+12Leak"].fillna(False, inplace=True)
data["t+16Leak"].fillna(False, inplace=True)
data.isnull().sum()
```

```python
data.shape[0]
```

```python
data.to_pickle("./trainingSet")
```
