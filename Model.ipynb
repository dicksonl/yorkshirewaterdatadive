```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_pickle("./northnf")
df['dev'] = df.Flow - df.intervalMean  
```

```python
df["t-4dev"] = df.groupby("DMA").dev.shift(-4)
df["t-8dev"] = df.groupby("DMA").dev.shift(-8)
df["t-12dev"] = df.groupby("DMA").dev.shift(-12)
df["t-16dev"] = df.groupby("DMA").dev.shift(-16)

df["t4Leak"] = df.groupby("DMA").exist.shift(4)
df["t8Leak"] = df.groupby("DMA").exist.shift(8)
df["t12Leak"] = df.groupby("DMA").exist.shift(12)
df["t16Leak"] = df.groupby("DMA").exist.shift(16)
```

```python
df.head()
```

```python
data = df[["DMA","Flow", "dev","t-4dev","t-8dev","t-12dev","t-16dev","t4Leak","t8Leak","t12Leak","t16Leak","exist"]]
data.head(100)
```

```python
data.isnull().sum()
```

```python
data.shape[0]
```

```python
data = data.dropna(subset=['t-4dev', 't-8dev', 't-12dev', 't-16dev'], how='any') 
data["t4Leak"].fillna(False, inplace=True)
data["t8Leak"].fillna(False, inplace=True)
data["t12Leak"].fillna(False, inplace=True)
data["t16Leak"].fillna(False, inplace=True)
data.isnull().sum()
```

```python
data.shape[0]
```

```python
data["LeakFixIn4Days"] = data[["exist", "t4Leak","t16Leak","t8Leak","t12Leak"]].any(1)
```

```python
data = data.drop(columns=['t4Leak', 't8Leak', 't12Leak', 't16Leak', 'exist'])
```
