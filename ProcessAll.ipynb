```python
df = pd.read_csv("./data/north.csv")
df['DMA'].unique()
```

```python
dma = df.drop(['Unnamed: 0','Area', 'Flow Validity Code','Number Of Household Properties', 'Number Of Non-Household Properties'], axis=1)
```

```python
dma.isnull().sum()
```

```python
dma['Datetime'] = pd.to_datetime(dma['Datetime'])
dma['hour'] = dma['Datetime'].dt.hour
dma['day_of_week'] = dma['Datetime'].dt.dayofweek
dma = dma.set_index('Datetime')
```

```python
dma['intervalMean'] = dma.groupby([dma.DMA, dma.day_of_week, dma.Time])['Flow'].transform(np.mean)
```

```python
empt = dma[dma.isnull().any(axis=1)]
```

```python
dma = dma[dma['DMA'] != 333]
dma = dma[dma['DMA'] != 826]
dma.Flow.fillna(dma.intervalMean, inplace=True)
empt = dma[dma.isnull().any(axis=1)]
empt.head()
```
